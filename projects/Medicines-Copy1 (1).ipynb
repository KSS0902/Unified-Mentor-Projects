```python
data1.fillna('', inplace=True)
```

```python
# Standardize data: lowercasing side effects and substitutes
side_effect_columns = [col for col in data1.columns if 'sideEffect' in col]
for col in side_effect_columns:
    data1[col] = data1[col].str.lower()
```

```python
# Convert therapeutic class to category
data1['Therapeutic Class'] = data1['Therapeutic Class'].astype('category')
```

```python
# Count unique substitutes for each medication
substitutes_columns = [col for col in data1.columns if 'substitute' in col]
substitute_counts = data1[substitutes_columns].apply(lambda x: x.nunique(), axis=1)
```

```python
# Frequency analysis of substitutes
substitute_frequency = pd.Series(data1[substitutes_columns].values.ravel()).value_counts()
print(substitute_frequency)
```

```python
# List unique side effects
unique_side_effects = pd.Series(data1[side_effect_columns].values.ravel()).unique()
print("Unique Side Effects:", unique_side_effects)
```

```python
# Frequency of side effects
side_effect_frequency = pd.Series(data1[side_effect_columns].values.ravel()).value_counts()
print(side_effect_frequency)
```

```python
# Analyze uses
uses = data1[['use0', 'use1', 'use2', 'use3', 'use4']].values.ravel()
uses_frequency = pd.Series(uses).value_counts()
print(uses_frequency)
```

```python
# Class distribution
class_distribution = data1['Therapeutic Class'].value_counts()
print(class_distribution)
```

```python
# Correlation Analysis
correlation_matrix = data1.corr()
print(correlation_matrix)
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Substitute Frequency Plot
plt.figure(figsize=(12, 6))
substitute_frequency.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Common Substitutes')
plt.xlabel('Substitutes')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
# Side Effect Frequency Plot
plt.figure(figsize=(12, 6))
side_effect_frequency.head(10).plot(kind='bar', color='lightcoral')
plt.title('Top 10 Most Common Side Effects')
plt.xlabel('Side Effects')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
# Uses Frequency Plot
plt.figure(figsize=(12, 6))
uses_frequency.head(10).plot(kind='bar', color='gold')
plt.title('Top 10 Uses of Medications')
plt.xlabel('Uses')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
# Class Distribution Pie Chart
plt.figure(figsize=(8, 8))
class_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Therapeutic Class Distribution')
plt.ylabel('')  # Hide the y-label
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(data1.corr(), annot=True, fmt='.2f', cmap='coolwarm', square=True, cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```
